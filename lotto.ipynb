```python
dataset = [
	['아메리카노', '카페라떼'],
	['카페라떼', '아메리카노', '카푸치노'],
	['바닐라라떼', '아메리카노'],
	['녹차라떼', '카페라떼', '아메리카노'],
	['카페모카', '아메리카노'],
	['아메리카노', '카페라떼'],
	['초콜릿', '아메리카노'],
	['아메리카노'],
	['카페모카', '카페라떼']
]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_result = te.fit(dataset).transform(dataset)
```

```python
te_result
```

```python
te.transform(dataset)
```

```python
import pandas as pd
```

```python
te
```

```python
te.columns_mapping_
```

```python
te.columns_
```

```python
df=pd.DataFrame(columns=te.columns_,data=te_result)
```

```python
df
```

```python
itemset = apriori(df, use_colnames=True)
itemset
```

```python
itemset = apriori(df, min_support=0.1, use_colnames=True)
itemset
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(itemset, metric="confidence",min_threshold=0.8) 
```

```python
lotto=pd.read_csv('lotto.csv')
```

```python
lotto
```

```python
lotto=lotto.drop(labels='time_id', axis=1)
```

```python
lotto
```

```python
lotto.values
```

```python
tre=TransactionEncoder()
```

```python
td=tre.fit(lotto.values).transform(lotto.values)
td
```

```python
td.shape
```

```python
dataf=pd.DataFrame(columns=tre.columns_,data=td)
dataf
```

```python
itemset = apriori(dataf, use_colnames=True, min_support=0.002, max_len=6)
#itemset['length']=itemset['itemsets'].apply(lambda x:len(x))
#item=itemset[itemset['length']>=2]
item=itemset
```

```python
result=association_rules(item, metric="confidence",min_threshold=0.8).sort_values(by='lift',ascending=False)
result
```
